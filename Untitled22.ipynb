```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('nonhelmet.jpg')

# Check if image is loaded properly
if image is None:
    print("Error: Image file not found or invalid.")
    exit()

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply a Gaussian blur to the image
blur = cv2.GaussianBlur(gray, (5, 5), 0)

# Threshold the image to get a binary image
thresh = cv2.threshold(blur, 120, 255, cv2.THRESH_BINARY)[1]

# Find contours in the binary image
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Check if any contours were found
if len(contours) == 0:
    print("Error: No contours found in the image.")
    exit()

# Find the largest contour
max_contour = max(contours, key=cv2.contourArea)

# Draw the contour on the image
cv2.drawContours(image, [max_contour], -1, (0, 255, 0), 2)

# Show the image using cv2_imshow
from google.colab.patches import cv2_imshow
cv2_imshow(image)

# Keep the window open until a key is pressed
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
!pip install pytesseract
import cv2
import pytesseract

# Read the image
image = cv2.imread('numberplate.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding to the image
thresh = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)[1]

# Find contours in the image
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# Find the largest contour
max_contour = max(contours, key=cv2.contourArea)

# Check if the contour has more than 0 points
if len(max_contour) > 0:
    # Check if the contour has more than 1 point in the first dimension
    if len(max_contour[0]) > 1:
        # Crop the image to the contour
        cropped_image = image[max_contour[0][0][1]:max_contour[0][2][1], max_contour[0][0][0]:max_contour[0][2][0]]

        # Apply OCR to the cropped image
        text = pytesseract.image_to_string(cropped_image, lang='eng', config='--psm 6')

        # Print the extracted text
        print(text)



```
